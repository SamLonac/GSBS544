```python
#create pluralize_gift function
def pluralize_gift(gift):
  """
  Returns plural of a noun

  Parameters
  ----------
  gift: str
    A noun

  Return
  ------
  str
    Plural version of gift
  """
  if gift[-1] == "y":
    gift = gift.replace("y", "ies")
  elif gift.find("oo") != -1:
    gift = gift.replace("oo", "ee")
  else:
    gift = gift + "s"

  return gift
```

```python
#create make_phrase function utilizing pluralize_gift
def make_phrase(num, num_word, item, verb, adjective, location):
  """
  Constructs a phrase for the 'Twelve Days of Christmas' song.

  Parameters
  ----------
  num : int
      The day number to generate the phrase for.
  num_word : str
      The order of the day number in the sequence of the 'Twelve Days of Christmas song' (e.g., "first", "second").
  item : str
      The gift item.
  verb : str, optional
      A verb related to the item.
  adjective : str, optional
      An adjective that describes the item.
  location : str, optional
      A location that the item resides in.

  Returns
  -------
  str
      The constructed phrase.
  """

  num_to_written = {
        1: 'one',
        2: 'two',
        3: 'three',
        4: 'four',
        5: 'five',
        6: 'six',
        7: 'seven',
        8: 'eight',
        9: 'nine',
        10: 'ten',
        11: 'eleven',
        12: 'twelve'
    }

  num_written = num_to_written.get(num, str(num))

  ## Step 1: Replace NAs with blank strings
  verb = "" if pd.isna(verb) else verb
  adjective = "" if pd.isna(adjective) else adjective
  location = "" if pd.isna(location) else location

  ## Step 2: If the day number is larger than 1, the gift items need pluralized!
  ### Hint: call the function you created above!

  if num > 1:
    item = pluralize_gift(item)

  ## Step 3: Figure out if a gift item starts with a vowel

  vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']
  starts_with_vowel = item[0] in vowels

  ## Step 4: For the first day, if the gift item starts with a vowel, replace the day with "an" and if the gift item does not start with a vowel, replace the day with "a" (e.g. a partridge in a pear tree). If it is not the first day, use just the number word (e.g. ten lords a leap)

  if num == 1:
    an_or_a_or_num = "an" if starts_with_vowel else "a"
  else:
    an_or_a_or_num = num_written
  ## Step 5: Put all of the pieces together into one string and return!

  #original code with spacing issue
  #phrase = (an_or_a_or_num + " " + adjective + " " + item + " " + verb + " " + location)
  #return phrase
  parts = [an_or_a_or_num, adjective, item, verb, location]
  phrase = ' '.join([part for part in parts if part])

  return phrase.strip()
```

```python
#create sing_day function utilizing previous functions
def sing_day(dataset, num, phrase_col):
  """
  Generates a string of the song 'Twelve Days of Christmas' for a specific day.

  Parameters
  ----------
  dataset : pd.DataFrame
      A DataFrame that contains the phrases for each of the gifts in the 'Twelve Days of Christmas' song.
  num : int
      The day number to generate the song string for.
  phrase_col : str
      The column name in 'dataset' where the phrases are located.

  Returns
  -------
  str
      A printed string of the song for the specific day.
  """
  num_to_order = {
    1: 'first',
    2: 'second',
    3: 'third',
    4: 'fourth',
    5: 'fifth',
    6: 'sixth',
    7: 'seventh',
    8: 'eighth',
    9: 'ninth',
    10: 'tenth',
    11: 'eleventh',
    12: 'twelfth'
    }

  # Step 1: Setup the intro line
  num_word = num_to_order[num]
  intro = "On the " + num_word + " day of Christmas, my true love sent to me:"

  # Step 2: Sing the gift phrases
  # Hint: What order are they gifts sung in each day?
  gifts = ""
  for i in range(num, 0, -1):
      if i == 1 and num != 1:
          gifts = gifts + "\nand " + dataset[phrase_col].iloc[i-1] + "." + "\n"
      elif i == 1 and num == 1:
          gifts = gifts + "\n" + dataset[phrase_col].iloc[i-1] + "\n"
      else:
          gifts = gifts + "\n" + dataset[phrase_col].iloc[i-1] + ","

# Step 3: Put it all together and return
  song = intro + gifts
  return print(song)
```

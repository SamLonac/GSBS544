## Does the distribution of each chain’s stores match population distribution?

```python
pop_data_region = merged_df.groupby(['Region']).agg({'Location Count': 'sum', 'Population': 'sum'}).reset_index()

pop_data_region
```

```python
#calculating the correlation coefficient between 'Location Count' and 'Population' by region
region_correlation_coefficient = pop_data_region['Location Count'].corr(pop_data_region['Population'])

region_correlation_coefficient
```

```python
pop_data_state = merged_df.groupby(['State']).agg({'Location Count': 'sum', 'Population': 'sum'}).reset_index()

pop_data_state
```

```python
#calculating the correlation coefficient between 'Location Count' and 'Population' by state
state_correlation_coefficient = pop_data_state['Location Count'].corr(pop_data_state['Population'])

state_correlation_coefficient
```

The correlation coefficients for location count and population for region and state were very strong, with a .981 and .942 respectively. This would indicate a strong linear relationship between the number of locations of coffee shops in a given state and the population of that state. This logically makes sense, as the more people in a given state, the more potential consumers of coffee.

```python
correlations = merged_df.groupby('Coffee Chain').apply(lambda x: x['Location Count'].corr(x['Population']))

correlations
```

When analyzing by chain, the correlation between the number of locations and population in a state differs widely. For the smaller chains, there does not seem to be a correlation. This would indicate that they still have a lot of room to grow in locations regardless of the size of each state. In other words, they don't have many locations anywhere regardless of population. As for the larger chains such as Starbucks, McDonald's and Dunkin', there is a strong correlation. This shows that they have a large coverage of the United States, focusing on expanding to a point where there is a location in any population center.

# Automate

Converting code to a function that takes a single argument(menuism url) and returns a clean, organized dataset with state abbreviation, location count, and company name columns.

```python
def automate_locations(url):
    rows = []
    pattern = r'.*/([^/]+)$'
    match = re.search(pattern, url)
    name_url = match.group(1) if match else None
    split_name = name_url.split("-")[:-1]
    chain_name = " ".join(split_name).title()

    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    results = soup.find_all("a", attrs={"class" : ""})
    state_pattern = fr"(.*)(?=\s+{chain_name} locations)"
    number_pattern= fr"(?<={chain_name} locations \()(\d+)"

    for result in results:
        state_match = re.search(state_pattern, result.text)
        number_match = re.search(number_pattern, result.text)
        if (state_match != None) & (number_match != None):
            state = state_match.group(0)
            number = int(number_match.group(1))

            rows.append({
            "State": state,
            "Location Count" : number,
            "Company Name" : chain_name
            })

        else:
            None

    return pd.DataFrame(rows)
```

```python
def recommend_trails(dataset, name, min_elev = None, min_rating = None, type = None):

  if min_elev is None:
    min_elev = 0
  if min_rating is None:
      min_rating = 0
  if type is None:
    type = 'loop'

  filtered_data = dataset[(dataset['area_name'] == name) & (dataset['elevation_gain'] > min_elev) & (dataset['difficulty_rating'] > min_rating) & (dataset['route_type'] == type)]

  return filtered_data
```

```python
recommend_trails(np_trails, "Haleakala National Park", min_elev = 1000, min_rating = 4)
```

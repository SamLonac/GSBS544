# Practice Activity 7.1

Consider four possible models for predicting house prices:

- Using only the size and number of rooms.
- Using size, number of rooms, and building type.
- Using size and building type, and their interaction.
- Using a 5-degree polynomial on size, a 5-degree polynomial on number of rooms, and also building type.

Set up a pipeline for each of these four models.

Then, get predictions on the test set for each of your pipelines, and compute the root mean squared error. Which model performed best?

Note: You should only use the function train_test_split() one time in your code; that is, we should be predicting on the same test set for all three models.

```python
X = ames.drop("SalePrice", axis = 1)
y = ames["SalePrice"]



X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 4)
```

```python
enc = OneHotEncoder(sparse_output = False)
interact = PolynomialFeatures(interaction_only = True)
do_nothing = FunctionTransformer(lambda x: x)
```

```python
X.columns
```

### Model 1: Using only the size and number of rooms

```python
ct = ColumnTransformer(
  [
    ("keep", do_nothing, ["Gr Liv Area", "TotRms AbvGrd"])
  ], remainder = "drop"
)


lr_pipeline_1 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

lr_fitted = lr_pipeline_1.fit(X_train, y_train)
y_preds_model_1 = lr_fitted.predict(X_test)
rmse_model_1 = mean_squared_error(y_test, y_preds_model_1, squared = False)
```

```python
rmse_model_1
```

### Model 2: Using size, number of rooms, and building type.

```python
ct = ColumnTransformer(
  [
    ("keep", do_nothing, ["Gr Liv Area", "TotRms AbvGrd"]),
    ("dummify", enc, ["Bldg Type"])

  ], remainder = "drop"
)


lr_pipeline_2 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

lr_fitted = lr_pipeline_2.fit(X_train, y_train)
y_preds_model_2 = lr_fitted.predict(X_test)
rmse_model_2 = mean_squared_error(y_test, y_preds_model_2, squared = False)
```

```python
rmse_model_2
```

### Model 3: Using size and building type, and their interaction

```python
ct_1 = ColumnTransformer(
  [
    ("keep", do_nothing, ["Gr Liv Area"]),
    ("dummify", enc, ["Bldg Type"])

  ], remainder = "drop"
)

ct_2 = ColumnTransformer(
  [
    ("inter_term_1", interact, ["dummify__Bldg Type_1Fam", "keep__Gr Liv Area"]),
    ("inter_term_2", interact, ["dummify__Bldg Type_2fmCon", "keep__Gr Liv Area"]),
    ("inter_term_3", interact, ["dummify__Bldg Type_Duplex", "keep__Gr Liv Area"]),
    ("inter_term_4", interact, ["dummify__Bldg Type_Twnhs", "keep__Gr Liv Area"]),
    ("inter_term_5", interact, ["dummify__Bldg Type_TwnhsE", "keep__Gr Liv Area"]),

  ], remainder = "drop"
)


lr_pipeline_3 = Pipeline(
  [("preprocessing_step_1", ct_1),
   ("preprocessing_step_2", ct_2),
  ("linear_regression", LinearRegression())]
).set_output(transform = "pandas")

lr_fitted = lr_pipeline_3.fit(X_train, y_train)
y_preds_model_3 = lr_fitted.predict(X_test)
rmse_model_3 = mean_squared_error(y_test, y_preds_model_3, squared = False)
```

```python
ct_1.transform(X_train).head()
```

```python
ct_2.transform(ct_1.transform(X_train)).head()
```

```python
rmse_model_3
```

### Model 4: Using a 5-degree polynomial on size, a 5-degree polynomial on number of rooms, and also building type.

```python
poly = PolynomialFeatures(5)
ct = ColumnTransformer(
  [
    ("dummify", enc, ["Bldg Type"]),
    ("deg5_size", poly, ["Gr Liv Area"]),
    ("deg5_rooms", poly, ["TotRms AbvGrd"])

  ], remainder = "drop"
)

lr_pipeline_4 = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
).set_output(transform = "pandas")

lr_fitted = lr_pipeline_4.fit(X_train, y_train)
y_preds_model_4 = lr_fitted.predict(X_test)
rmse_model_4 = mean_squared_error(y_test, y_preds_model_4, squared = False)
```

```python
ct.transform(X_train)
```

```python
rmse_model_4
```

Model 3 is the best model with the lowest RMSE of 51,378.

### Step 3: GridSearch

```python
ct = ColumnTransformer(
    [
        ("dummify", enc, ["Bldg Type"]),
        ("polynomial_size", PolynomialFeatures(), ["Gr Liv Area"]),
        ("polynomial_rooms", PolynomialFeatures(), ["TotRms AbvGrd"])
    ], remainder="drop"
)

lr_pipeline_5 = Pipeline(
    [("preprocessing", ct),
     ("linear_regression", LinearRegression())]
).set_output(transform = "pandas")

degrees = {
    'preprocessing__polynomial_size__degree': np.arange(1, 11),
    'preprocessing__polynomial_rooms__degree': np.arange(1, 11)
}

gscv = GridSearchCV(lr_pipeline_5, degrees, cv = 5, scoring='r2')
gscv_fitted = gscv.fit(X, y)
gscv_fitted.cv_results_['mean_test_score']
```

```python
# Note: The following fit operation is commented out as it requires the actual dataset.
# gscv_updated.fit(X, y)

# After fitting, we can extract the results
results_df = pd.DataFrame(gscv_fitted.cv_results_)
results_df = results_df[['param_preprocessing__polynomial_size__degree', 'param_preprocessing__polynomial_rooms__degree', 'mean_test_score', 'rank_test_score']]
results_df = results_df.rename(columns={
     'param_preprocessing__polynomial_size__degree': 'house size polynomial',
     'param_preprocessing__polynomial_rooms__degree': 'rooms polynomial',
     'mean_test_score': 'Score'
 })
results_df.sort_values('rank_test_score')
```

The model using a 3-degree polynomial on size, a 1-degree polynomial on number of rooms, and also building type performed this best.

The issue with trying all possible model options is that it could take excess time and resources to find a model that scores the best, but may just be overfitting the data. With high number polynomial transformations there is a greater chance that the data is being overfit.
